# Reading *.csv data

```python
import pandas as pd
from os import path
from glob import glob
```

```python
DATA_PATH = path.join(path.pardir, "data")
TEST_FILE_PATH = path.join(DATA_PATH, "data.csv")
```

```python
df = pd.read_csv(TEST_FILE_PATH)
```

```python
df.head(10)
```

```python
COLUMNS_TO_DROP = ["Wahlperiode", 
                   "Sitzungnr", 
                   "Name", 
                   "Vorname",
                   "Titel",
                   "Enthaltung", 
                   "ungültig", 
                   "nichtabgegeben", 
                   "Bemerkung"]

df.drop(COLUMNS_TO_DROP, axis=1, inplace=True)
```

```python
df.head(10)
```

```python
df["Abstimmung"] = "data.csv"
df["vote"] = df["ja"] - df["nein"]
df.drop(["ja", "nein"], axis=1, inplace=True)
```

```python
df.head(10)
```
